meta if else

```c++
template <bool condition,typename THEN,typename ELSE>
struct if_ ;

template <typename THEN,typename ELSE>
struct if_<true,THEN,ELSE>{
    using type = THEN;
};

template <typename THEN,typename ELSE>
struct if_<false,THEN,ELSE>{
    using type = ELSE;
};

```

```c++
std::string s = typeid(if_<false,int,float>::type).name();
std::cout << s <<std::endl;
```
