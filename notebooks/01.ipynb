meta if else

```c++

template <bool condition,typename THEN,typename ELSE>
struct if_ ;

template <typename THEN,typename ELSE>
struct if_<true,THEN,ELSE>{
    using type = THEN;
};

template <typename THEN,typename ELSE>
struct if_<false,THEN,ELSE>{
    using type = ELSE;
};

```

```c++
std::string s = typeid(if_<false,int,float>::type).name();
std::cout << s <<std::endl;
```

variadic template

```c++
void printn(){
    std::cout << std::endl;
};
```

```c++
template <typename LAST>
void printn(LAST t){
    std::cout << t << std::endl;
}
template <typename T0, typename ... T1toN>
void printn(T0 t0,T1toN... rest){
   std::cout << t0 << " , ";
   printn(rest...);
}
```

```c++
printn(1,"hello",2.0f);
```
